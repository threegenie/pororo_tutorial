```python
df = pd.read_table('naver_shopping.txt', names=['ratings', 'reviews'])
df
```

```python
sa = Pororo(task='sentiment',lang='ko')
```

```python
df['sentiment'] = None
```

```python
df.head()
```

```python
data = df[:50] #데이터가 너무 크니까 50개만 보기
```

```python
for i in range(len(data['reviews'])):
  data['sentiment'][i] = sa(data['reviews'][i])
```
